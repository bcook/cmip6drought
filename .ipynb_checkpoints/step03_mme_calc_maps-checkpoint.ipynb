# Generate figures for an MME ensemble.
## How it Works:
* Choose variable
* Average all ensemble members within a model first
* Then average together into MME
* Then generate MME figure in units of percent change
* Database here: http://mary.ldeo.columbia.edu:81/CMIP6/
* For master list of everything available: wget http://mary.ldeo.columbia.edu/master_collection/mary_cmip6.csv

## Generate list of models

```python
# Number of variables required
numvar_req = np.size(var_list)

# Load files
varlist_ssp  = pd.read_csv('../varlists/model_procvar_'+ssp+'.csv'); 
varlist_hist = pd.read_csv('../varlists/model_procvar_hist.csv'); 

# Pull out arrays of all possible models/ensemble members
list_allmodels = np.array(varlist_ssp.Model)
list_allensemb = np.array(varlist_ssp.ensemble)

# Pull out array of all column headers
varlist_header=np.array(varlist_ssp.columns)

# Loop through and find the location of all these variables in the dataframe
loc_vars = np.zeros(np.shape(var_list))*np.nan
for i_var in enumerate(var_list):

    # Find the column for the current variable and store this information
    loc_vars[i_var[0]] = np.where(varlist_header==i_var[1])[0]
    
# Calculate the number of variables available for each model/ensemble run
numvar_avail_ssp  = np.sum(np.array(varlist_ssp.iloc[:,loc_vars]),axis=1)
numvar_avail_hist = np.sum(np.array(varlist_hist.iloc[:,loc_vars]),axis=1)

# Find simulations where all these variables are available
i_mods = np.where( (numvar_avail_ssp==numvar_req) & (numvar_avail_hist==numvar_req) )[0]

# Pull out list of models and ensemble designations
valid_models = list_allmodels[i_mods].copy()
valid_ensemb = list_allensemb[i_mods].copy()

unique_models = np.unique(valid_models)

# shortcut already processed models
unique_models=unique_models[:]

print(unique_models)
```

## Compile Multi-Model Ensemble

```python
# Input Directory
#input_dir = '/Volumes/Apollo/CMIP6/cmip6interp/'
input_dir = '/Volumes/Raven/CMIP6/cmip6interp/'

# Storage Arrays
allmods_hist = np.zeros((np.size(unique_models),yrs_hist.size,mons.size,lat_intp.size,lon_intp.size))*np.nan
allmods_ssp  = np.zeros((np.size(unique_models),yrs_ssp.size,mons.size,lat_intp.size,lon_intp.size))*np.nan

# Model Loop
for n_model in enumerate(unique_models):

    # Load All data from current model
    nc_link = input_dir+curr_var+'.regrid.'+n_model[1]+'.hist.'+ssp+'.1850.2100.nc'               
    print(nc_link)
   
    # Load year of data
    nc_file = netCDF4.Dataset(nc_link) 
    
    # Pull out dimension variables 
    yr     = nc_file.variables['yr'][:]
    lat    = nc_file.variables['lat'][:]
    lon    = nc_file.variables['lon'][:]
      
    # Find Years
    iyrs_hist = np.nonzero(np.in1d(yr,yrs_hist))[0]
    iyrs_ssp  = np.nonzero(np.in1d(yr,yrs_ssp))[0]
    
    # Load all ensemble members for this period
    allensemb_hist = nc_file.variables[curr_var][:,iyrs_hist,:,:,:]
    allensemb_ssp  = nc_file.variables[curr_var][:,iyrs_ssp,:,:,:]

    # Ensemble Average
    currmod_hist = np.mean(allensemb_hist,axis=0)
    currmod_ssp  = np.mean(allensemb_ssp,axis=0)

    # Store in MME array
    allmods_hist[n_model[0]] = currmod_hist
    allmods_ssp[n_model[0]]  = currmod_ssp
    
    # close file
    nc_file.close()
    
```

## Create Land-Sea Mask Using a Land Hydrology File and COSINE Area Mask

```python
# Load lat/lon from regridded file to create land sea mask, composited from all models
#nc_link = '/Volumes/Apollo/CMIP6/cmip6interp/mrsos.regrid.BCC-CSM2-MR.hist.ssp585.1850.2100.nc'
#nc_link = '/Volumes/Raven/CMIP6/cmip6interp/mrsos.regrid.BCC-CSM2-MR.hist.ssp585.1850.2100.nc'
#nc_link = '/Volumes/Raven/CMIP6/cmip6interp/mrsos.regrid.CESM2.hist.ssp585.1850.2100.nc'

# Store land-sea masks from each model
lsmask_all = np.zeros((np.size(unique_models),np.size(lat),np.size(lon)))*np.nan

# Model Loop
for n_model in enumerate(unique_models):
    nc_link = '/Volumes/Raven/CMIP6/cmip6interp/mrsos.regrid.'+n_model[1]+'.hist.ssp585.1850.2100.nc'
    nc_file = netCDF4.Dataset(nc_link) 
    
    # Pull out variables 
    lat          = nc_file.variables['lat'][:]
    lon          = nc_file.variables['lon'][:]
    lsmask_model = nc_file.variables['mrsos'][0,0,0,:,:]

    lsmask_model[np.where(lsmask_model>9e10)]=0
    lsmask_model[np.where(lsmask_model>0)]=1
    lsmask_model[np.where(lsmask_model<=0)]=0
    
    lsmask_all[n_model[0],:,:]=lsmask_model
    
# Create Ls Mask
lsmask = np.sum(lsmask_all,axis=0)
lsmask[np.where(lsmask<np.size(unique_models))]=0
lsmask[np.where(lsmask>0)]=1

# Filter out Antarctica and Greenland
#lsmask[np.where(lat<-62),:]=0
#i_lat_greenland = np.where( (lat>=60) )[0]
#i_lon_greenland = np.where( (lon>=-50) & (lon<=0) )[0]
#lsmask[i_lat_greenland,:][:,i_lon_greenland]=0
#lsmask[i_lat_greenland,:]=0

# close file
nc_file.close()

plt.figure(facecolor="white")
plt.pcolormesh(lon,lat,lsmask)
plt.colorbar()

# Create Latitude Weights
lat_wts = scipy.cos(scipy.deg2rad(lat));
lat_wts_grid,lon_junk = np.meshgrid(lat_wts,lon)
lat_wts_grid=np.swapaxes(lat_wts_grid,1,0)   #*reg_area

plt.figure(facecolor="white")
plt.pcolormesh(lon,lat,lsmask*lat_wts_grid)
plt.colorbar()
```

## Seasonal Difference: DJF

```python
# Find Land Area With Robust Changes
landarea_mask = lsmask*lat_wts_grid

# For CDF histograms for robustness calculation
num_bins = 1000

# Setup
i_seas = np.nonzero(np.in1d(mons,np.array([1,2,12])))[0]; seas_txt='DJF'
#i_seas = np.nonzero(np.in1d(mons,np.array([6,7,8])))[0]; seas_txt='JJA'
#i_seas = np.nonzero(np.in1d(mons,np.array([3,4,5])))[0]; seas_txt='MAM'
#i_seas = np.nonzero(np.in1d(mons,np.array([9,10,11])))[0]; seas_txt='SON'

# Calculate seasonal average
allmods_seas_hist = np.mean(allmods_hist[:,:,i_seas,:,:],axis=2)
allmods_seas_ssp  = np.mean(allmods_ssp[:,:,i_seas,:,:],axis=2)

# Array to store robustness metric
Rrob = np.zeros((lat.size,lon.size))*np.nan

# Loop Through each grid cell and calculate
for n_lat in enumerate(lat):
    for n_lon in enumerate(lon):
               
        # Ignore Ocean Cells
        if lsmask[n_lat[0],n_lon[0]]>0:
        
            #if np.max(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]])<80000:
            if landarea_mask[n_lat[0],n_lon[0]]>=0:

                # A1 Calculation (based on MME and full ensemble DIFFERENCES)-----------------------------------------------------------------
                # difference, full ensemble
                currcell_allmods = np.ndarray.flatten(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]])
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                # difference, MME
                currcell_mme     = np.mean( (allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]]),axis=0   )
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                
                # Create normalized CDFs
                # ensemble
                counts, bin_edges = np.histogram(currcell_allmods, bins=num_bins, normed=True)
                cdf_ens      = np.cumsum(counts)
                cdf_ens_norm = cdf_ens/cdf_ens[-1]
                # mme
                counts, bin_edges = np.histogram(currcell_mme, bins=num_bins, range=(np.min(bin_edges),np.max(bin_edges)), normed=True)
                cdf_mme      = np.cumsum(counts)
                cdf_mme_norm = cdf_mme/cdf_mme[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_ens/cdf_ens[-1])
                #plt.plot(bin_edges[1:], cdf_mme/cdf_mme[-1],color='r')

                # Calculate A1
                a1=np.sum((cdf_ens_norm-cdf_mme_norm)**2)
  
                # A2 Calculation----------------------------------------------------------------------------------------------------------
                # MME, historical
                currcell_mme_hist = np.ndarray.flatten(np.mean(allmods_seas_hist[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_hist,index=None)).to_csv('hist_mmm.csv',index=False)
                # MME, SSP
                currcell_mme_ssp  = np.ndarray.flatten(np.mean(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_ssp,index=None)).to_csv('fut_mmm.csv',index=False)

                # Ranges for CDFs
                bin_min=np.min((np.min(currcell_mme_hist),np.min(currcell_mme_ssp)))
                bin_max=np.max((np.max(currcell_mme_hist),np.max(currcell_mme_ssp)))

                # Historical MME CDF
                counts, bin_edges = np.histogram(currcell_mme_hist, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_hist      = np.cumsum(counts)
                cdf_hist_norm = cdf_hist/cdf_hist[-1]
                # SSP MME CDF
                counts, bin_edges = np.histogram(currcell_mme_ssp, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_ssp       = np.cumsum(counts)
                cdf_ssp_norm  = cdf_ssp/cdf_ssp[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_hist/cdf_hist[-1])
                #plt.plot(bin_edges[1:], cdf_ssp/cdf_ssp[-1],color='r')

                # Calculate A2
                a2=np.sum((cdf_hist_norm-cdf_ssp_norm)**2)

                # Calculate Robustness Metric
                Rrob[n_lat[0],n_lon[0]] = 1-(a1/a2)

# Create MME Figure (percent units)-------------------------------------------------------------------------------------------------------
txt_title=curr_var+':'+ssp+' (%)'

# MMME Means
mme_mean_hist = np.mean(np.mean(allmods_seas_hist,axis=0),axis=0)
mme_mean_ssp  = np.mean(np.mean(allmods_seas_ssp,axis=0),axis=0)

# Create difference map (percent change)
anom_field=((mme_mean_ssp-mme_mean_hist)/mme_mean_hist)*100

# Mask ocean/ice sheet areas
anom_field[np.where(lsmask==0)]=np.nan

# Significance masking
nonsiglat,nonsiglon= np.where((Rrob)>=0.9)
latmask=lat[nonsiglat].copy()
lonmask=lon[nonsiglon].copy()

fig = plt.figure(figsize=(15, 7),facecolor='white')
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=0))
m = ax.contourf(lon, lat, anom_field, clevs*100, transform=ccrs.PlateCarree(),cmap=nmap,extend="both")
ax.coastlines()
ax.set_global()
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=1)
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=0,edgecolor='k')
ax.add_feature(cartopy.feature.LAND,zorder=1,facecolor='none',edgecolor='k',linewidth=0.5)
#ax.text(5,-55,model_name,transform=ccrs.PlateCarree(),fontsize=24,fontweight="bold")
## Significance Masking
plot_point(ax,lonmask,latmask)
ax.text(-145,-42,seas_txt,transform=ccrs.PlateCarree(),fontsize=28,fontweight="bold", \
        horizontalalignment='left', verticalalignment='center',)
# Add a Rectangle and some annotation
#for n_reg in enumerate(region_names):
#    plot_rectangle(ax,latlon_all[n_reg[0],0],latlon_all[n_reg[0],1],latlon_all[n_reg[0],2],latlon_all[n_reg[0],3])
plt.title(txt_title,fontsize=24)
cbar=plt.colorbar(m,orientation="horizontal",fraction=0.08,pad=0.04,ticks=clevs*100)
cbar.ax.tick_params(labelsize=24)
plt.show()
fig.savefig('../figures/'+curr_var+'.'+ssp+'.'+seas_txt+'.eps',format='eps')

# Find Land Area With Robust Changes
landarea_mask = lsmask*lat_wts_grid

# Fractional areas
fracrob_dryarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field<0))])/np.sum(landarea_mask)
print("Land Area, Drying: "+np.str(fracrob_dryarea))
fracrob_wetarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field>0))])/np.sum(landarea_mask)
print("Land Area, Wetting: "+np.str(fracrob_wetarea))
frac_nochange = np.sum(landarea_mask[ np.where(Rrob*lsmask<0.9)])/np.sum(landarea_mask)
print("Land Area, No Robust Change: "+np.str(frac_nochange))
print("SUM TOTAL: "+np.str(fracrob_dryarea+fracrob_wetarea+frac_nochange))
```

```python
# Find Land Area With Robust Changes
landarea_mask = lsmask*lat_wts_grid

# For CDF histograms for robustness calculation
num_bins = 1000

# Setup
#i_seas = np.nonzero(np.in1d(mons,np.array([1,2,12])))[0]; seas_txt='DJF'
#i_seas = np.nonzero(np.in1d(mons,np.array([6,7,8])))[0]; seas_txt='JJA'
i_seas = np.nonzero(np.in1d(mons,np.array([3,4,5])))[0]; seas_txt='MAM'
#i_seas = np.nonzero(np.in1d(mons,np.array([9,10,11])))[0]; seas_txt='SON'

# Calculate seasonal average
allmods_seas_hist = np.mean(allmods_hist[:,:,i_seas,:,:],axis=2)
allmods_seas_ssp  = np.mean(allmods_ssp[:,:,i_seas,:,:],axis=2)

# Array to store robustness metric
Rrob = np.zeros((lat.size,lon.size))*np.nan

# Loop Through each grid cell and calculate
for n_lat in enumerate(lat):
    for n_lon in enumerate(lon):
               
        # Ignore Ocean Cells
        if lsmask[n_lat[0],n_lon[0]]>0:
        
            #if np.max(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]])<80000:
            if landarea_mask[n_lat[0],n_lon[0]]>=0:

                # A1 Calculation (based on MME and full ensemble DIFFERENCES)-----------------------------------------------------------------
                # difference, full ensemble
                currcell_allmods = np.ndarray.flatten(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]])
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                # difference, MME
                currcell_mme     = np.mean( (allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]]),axis=0   )
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                
                # Create normalized CDFs
                # ensemble
                counts, bin_edges = np.histogram(currcell_allmods, bins=num_bins, normed=True)
                cdf_ens      = np.cumsum(counts)
                cdf_ens_norm = cdf_ens/cdf_ens[-1]
                # mme
                counts, bin_edges = np.histogram(currcell_mme, bins=num_bins, range=(np.min(bin_edges),np.max(bin_edges)), normed=True)
                cdf_mme      = np.cumsum(counts)
                cdf_mme_norm = cdf_mme/cdf_mme[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_ens/cdf_ens[-1])
                #plt.plot(bin_edges[1:], cdf_mme/cdf_mme[-1],color='r')

                # Calculate A1
                a1=np.sum((cdf_ens_norm-cdf_mme_norm)**2)
  
                # A2 Calculation----------------------------------------------------------------------------------------------------------
                # MME, historical
                currcell_mme_hist = np.ndarray.flatten(np.mean(allmods_seas_hist[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_hist,index=None)).to_csv('hist_mmm.csv',index=False)
                # MME, SSP
                currcell_mme_ssp  = np.ndarray.flatten(np.mean(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_ssp,index=None)).to_csv('fut_mmm.csv',index=False)

                # Ranges for CDFs
                bin_min=np.min((np.min(currcell_mme_hist),np.min(currcell_mme_ssp)))
                bin_max=np.max((np.max(currcell_mme_hist),np.max(currcell_mme_ssp)))

                # Historical MME CDF
                counts, bin_edges = np.histogram(currcell_mme_hist, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_hist      = np.cumsum(counts)
                cdf_hist_norm = cdf_hist/cdf_hist[-1]
                # SSP MME CDF
                counts, bin_edges = np.histogram(currcell_mme_ssp, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_ssp       = np.cumsum(counts)
                cdf_ssp_norm  = cdf_ssp/cdf_ssp[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_hist/cdf_hist[-1])
                #plt.plot(bin_edges[1:], cdf_ssp/cdf_ssp[-1],color='r')

                # Calculate A2
                a2=np.sum((cdf_hist_norm-cdf_ssp_norm)**2)

                # Calculate Robustness Metric
                Rrob[n_lat[0],n_lon[0]] = 1-(a1/a2)

# Create MME Figure (percent units)-------------------------------------------------------------------------------------------------------
txt_title=curr_var+':'+ssp+' (%)'

# MMME Means
mme_mean_hist = np.mean(np.mean(allmods_seas_hist,axis=0),axis=0)
mme_mean_ssp  = np.mean(np.mean(allmods_seas_ssp,axis=0),axis=0)

# Create difference map (percent change)
anom_field=((mme_mean_ssp-mme_mean_hist)/mme_mean_hist)*100

# Mask ocean/ice sheet areas
anom_field[np.where(lsmask==0)]=np.nan

# Significance masking
nonsiglat,nonsiglon= np.where((Rrob)>=0.9)
latmask=lat[nonsiglat].copy()
lonmask=lon[nonsiglon].copy()

fig = plt.figure(figsize=(15, 7),facecolor='white')
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=0))
m = ax.contourf(lon, lat, anom_field, clevs*100, transform=ccrs.PlateCarree(),cmap=nmap,extend="both")
ax.coastlines()
ax.set_global()
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=1)
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=0,edgecolor='k')
ax.add_feature(cartopy.feature.LAND,zorder=1,facecolor='none',edgecolor='k',linewidth=0.5)
#ax.text(5,-55,model_name,transform=ccrs.PlateCarree(),fontsize=24,fontweight="bold")
## Significance Masking
plot_point(ax,lonmask,latmask)
ax.text(-145,-42,seas_txt,transform=ccrs.PlateCarree(),fontsize=28,fontweight="bold", \
        horizontalalignment='left', verticalalignment='center',)
# Add a Rectangle and some annotation
#for n_reg in enumerate(region_names):
#    plot_rectangle(ax,latlon_all[n_reg[0],0],latlon_all[n_reg[0],1],latlon_all[n_reg[0],2],latlon_all[n_reg[0],3])
plt.title(txt_title,fontsize=24)
cbar=plt.colorbar(m,orientation="horizontal",fraction=0.08,pad=0.04,ticks=clevs*100)
cbar.ax.tick_params(labelsize=24)
plt.show()
fig.savefig('../figures/'+curr_var+'.'+ssp+'.'+seas_txt+'.eps',format='eps')

# Find Land Area With Robust Changes
landarea_mask = lsmask*lat_wts_grid

# Fractional areas
fracrob_dryarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field<0))])/np.sum(landarea_mask)
print("Land Area, Drying: "+np.str(fracrob_dryarea))
fracrob_wetarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field>0))])/np.sum(landarea_mask)
print("Land Area, Wetting: "+np.str(fracrob_wetarea))
frac_nochange = np.sum(landarea_mask[ np.where(Rrob*lsmask<0.9)])/np.sum(landarea_mask)
print("Land Area, No Robust Change: "+np.str(frac_nochange))
print("SUM TOTAL: "+np.str(fracrob_dryarea+fracrob_wetarea+frac_nochange))
```

## Seasonal Difference: AMJJAS

```python
# For CDF histograms for robustness calculation
num_bins = 1000

# Setup
i_seas = np.nonzero(np.in1d(mons,np.array([4,5,6,7,8,9])))[0]; seas_txt='AMJJAS'
#i_seas = np.nonzero(np.in1d(mons,np.array([1,2,3,10,11,12])))[0]; seas_txt='ONDJFM'

# Calculate seasonal average
allmods_seas_hist = np.mean(allmods_hist[:,:,i_seas,:,:],axis=2)
allmods_seas_ssp  = np.mean(allmods_ssp[:,:,i_seas,:,:],axis=2)

# Array to store robustness metric
Rrob = np.zeros((lat.size,lon.size))*np.nan

# Loop Through each grid cell and calculate
for n_lat in enumerate(lat):
    for n_lon in enumerate(lon):
               
        # Ignore Ocean Cells
        if lsmask[n_lat[0],n_lon[0]]>0:
        
            #if np.max(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]])<80000:
            if landarea_mask[n_lat[0],n_lon[0]]>=0:

                # A1 Calculation (based on MME and full ensemble DIFFERENCES)-----------------------------------------------------------------
                # difference, full ensemble
                currcell_allmods = np.ndarray.flatten(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]])
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                # difference, MME
                currcell_mme     = np.mean( (allmods_seas_ssp[:,:,n_lat[0],n_lon[0]]-allmods_seas_hist[:,:,n_lat[0],n_lon[0]]),axis=0   )
                # (pd.DataFrame(data=currcell_allmods,index=None)).to_csv('projchanges_allmodels.csv',index=False)
                
                # Create normalized CDFs
                # ensemble
                counts, bin_edges = np.histogram(currcell_allmods, bins=num_bins, normed=True)
                cdf_ens      = np.cumsum(counts)
                cdf_ens_norm = cdf_ens/cdf_ens[-1]
                # mme
                counts, bin_edges = np.histogram(currcell_mme, bins=num_bins, range=(np.min(bin_edges),np.max(bin_edges)), normed=True)
                cdf_mme      = np.cumsum(counts)
                cdf_mme_norm = cdf_mme/cdf_mme[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_ens/cdf_ens[-1])
                #plt.plot(bin_edges[1:], cdf_mme/cdf_mme[-1],color='r')

                # Calculate A1
                a1=np.sum((cdf_ens_norm-cdf_mme_norm)**2)
  
                # A2 Calculation----------------------------------------------------------------------------------------------------------
                # MME, historical
                currcell_mme_hist = np.ndarray.flatten(np.mean(allmods_seas_hist[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_hist,index=None)).to_csv('hist_mmm.csv',index=False)
                # MME, SSP
                currcell_mme_ssp  = np.ndarray.flatten(np.mean(allmods_seas_ssp[:,:,n_lat[0],n_lon[0]],axis=0))
                # (pd.DataFrame(data=currcell_mme_ssp,index=None)).to_csv('fut_mmm.csv',index=False)

                # Ranges for CDFs
                bin_min=np.min((np.min(currcell_mme_hist),np.min(currcell_mme_ssp)))
                bin_max=np.max((np.max(currcell_mme_hist),np.max(currcell_mme_ssp)))

                # Historical MME CDF
                counts, bin_edges = np.histogram(currcell_mme_hist, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_hist      = np.cumsum(counts)
                cdf_hist_norm = cdf_hist/cdf_hist[-1]
                # SSP MME CDF
                counts, bin_edges = np.histogram(currcell_mme_ssp, bins=num_bins, range=(bin_min,bin_max), normed=True)
                cdf_ssp       = np.cumsum(counts)
                cdf_ssp_norm  = cdf_ssp/cdf_ssp[-1]

                #plt.figure(facecolor="white")
                #plt.plot(bin_edges[1:], cdf_hist/cdf_hist[-1])
                #plt.plot(bin_edges[1:], cdf_ssp/cdf_ssp[-1],color='r')

                # Calculate A2
                a2=np.sum((cdf_hist_norm-cdf_ssp_norm)**2)

                # Calculate Robustness Metric
                Rrob[n_lat[0],n_lon[0]] = 1-(a1/a2)

# Create MME Figure (percent units)-------------------------------------------------------------------------------------------------------
txt_title=curr_var+':'+ssp+' (%)'

# MMME Means
mme_mean_hist = np.mean(np.mean(allmods_seas_hist,axis=0),axis=0)
mme_mean_ssp  = np.mean(np.mean(allmods_seas_ssp,axis=0),axis=0)

# Create difference map (percent change)
anom_field=((mme_mean_ssp-mme_mean_hist)/mme_mean_hist)*100

# Mask ocean/ice sheet areas
anom_field[np.where(lsmask==0)]=np.nan

# Significance masking
nonsiglat,nonsiglon= np.where((Rrob)>=0.9)
latmask=lat[nonsiglat].copy()
lonmask=lon[nonsiglon].copy()

fig = plt.figure(figsize=(15, 7),facecolor='white')
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=0))
m = ax.contourf(lon, lat, anom_field, clevs*100, transform=ccrs.PlateCarree(),cmap=nmap,extend="both")
ax.coastlines()
ax.set_global()
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=1)
ax.add_feature(cartopy.feature.OCEAN,color=ocean_color,zorder=0,edgecolor='k')
ax.add_feature(cartopy.feature.LAND,zorder=1,facecolor='none',edgecolor='k',linewidth=0.5)
#ax.text(5,-55,model_name,transform=ccrs.PlateCarree(),fontsize=24,fontweight="bold")
## Significance Masking
plot_point(ax,lonmask,latmask)
ax.text(-150,-42,seas_txt,transform=ccrs.PlateCarree(),fontsize=28,fontweight="bold", \
        horizontalalignment='left', verticalalignment='center',)
# Add a Rectangle and some annotation
#for n_reg in enumerate(region_names):
#    plot_rectangle(ax,latlon_all[n_reg[0],0],latlon_all[n_reg[0],1],latlon_all[n_reg[0],2],latlon_all[n_reg[0],3])
plt.title(txt_title,fontsize=24)
cbar=plt.colorbar(m,orientation="horizontal",fraction=0.08,pad=0.04,ticks=clevs*100)
cbar.ax.tick_params(labelsize=24)
plt.show()
fig.savefig('../figures/'+curr_var+'.'+ssp+'.'+seas_txt+'.eps',format='eps')

# Find Land Area With Robust Changes
landarea_mask = lsmask*lat_wts_grid

# Fractional areas
fracrob_dryarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field<0))])/np.sum(landarea_mask)
print("Land Area, Drying: "+np.str(fracrob_dryarea))
fracrob_wetarea = np.sum(landarea_mask[ np.where((Rrob*lsmask>=0.9) & (anom_field>0))])/np.sum(landarea_mask)
print("Land Area, Wetting: "+np.str(fracrob_wetarea))
frac_nochange = np.sum(landarea_mask[ np.where(Rrob*lsmask<0.9)])/np.sum(landarea_mask)
print("Land Area, No Robust Change: "+np.str(frac_nochange))
print("SUM TOTAL: "+np.str(fracrob_dryarea+fracrob_wetarea+frac_nochange))
```
